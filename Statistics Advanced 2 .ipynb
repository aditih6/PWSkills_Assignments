1.   What is hypothesis testing in statistics?
-  Hypothesis testing is a statistical method used to decide, based on sample data, whether there is enough evidence to reject a default assumption (the null hypothesis) about a population.
-  Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It evaluates whether the observed results are consistent with a stated assumption (null hypothesis) or provide enough evidence to support an alternative claim (alternative hypothesis).


Two hypotheses:
-  Null hypothesis (H₀): The default or “no change/no difference” assumption.
-  Alternative hypothesis (H₁): The claim that there is a difference or effect.

Significance level (α) - Chosen before the test (commonly 0.05), it represents the maximum probability of making a Type I error (rejecting H₀ when it’s true).

Test statistic - A value calculated from the sample data that measures how far the result is from what H₀ predicts. Examples: z-test, t-test, chi-square test.

p-value - The probability of obtaining results as extreme or more extreme than the observed data if H₀ were true.

Decision rule:
-  If p-value ≤ α → reject H₀ (evidence supports H₁).
-  If p-value > α → fail to reject H₀ (insufficient evidence against H₀).

Types of errors:
-  Type I error: Rejecting a true H₀ (false positive).
-  Type II error: Failing to reject a false H₀ (false negative).

2.   What is the null hypothesis, and how does it differ from the alternative hypothesis?
The alternative hypothesis is the statement that contradicts the null hypothesis. It represents the researcher’s claim or the effect/difference/relationship they want to detect.

Example: “The average height of male students in a college is not 170 cm.”
-  Key Point:
Hypothesis testing always begins with assuming H₀ is true. The data collected is then used to determine whether there is enough evidence to reject it in favor of H₁.

3.    Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The significance level (denoted by α) is the threshold probability set by the researcher before conducting the test. It represents the maximum acceptable risk of making a Type I error — rejecting the null hypothesis (H₀) when it is actually true.


Example: If α = 0.05, it means you are willing to accept a 5% chance of wrongly rejecting H₀.

The significance level is crucial because it determines how strong the evidence must be before rejecting H₀.

Steps in context:
-  Choose α (commonly 0.05, 0.01, or 0.10).
-  Collect sample data and calculate a test statistic (e.g., t, z, χ²).
-  Find the p-value.

Decision rule:
-  If p-value ≤ α -->> Reject H₀ (evidence supports H₁)
-  If p-value > α -->> Fail to reject H₀ (insufficient evidence)

Interpretation:
-  Small α (e.g., 0.01): Very strict — you need strong evidence to reject H₀ (reduces Type I errors but increases Type II errors).
-  Large α (e.g., 0.10): Less strict — easier to reject H₀ (increases risk of Type I errors).


Example: A factory claims its bulbs last 1000 hours. You test 50 bulbs.
-  H₀: Average life = 1000 hours
-  H₁: Average life ≠ 1000 hours
-  α = 0.05
-  Test gives p-value = 0.03

Since 0.03 ≤ 0.05, you reject H₀ and conclude the bulbs’ lifetime is likely different from 1000 hours.


Key Point:
The significance level does not measure the probability that H₀ is true or false — it only sets the decision threshold for rejecting H₀ based on the evidence.

4.  What are Type I and Type II errors? Give examples of each.


When performing hypothesis testing, there are two possible kinds of mistakes:
-  Type I Error (α error): Rejecting the null hypothesis (H₀) when it is actually true
-  Type II Error (β error): Failing to reject the null hypothesis when it is actually false

Type I Error Example:
- A medical test shows a patient has a disease when they are actually healthy
- H₀: Patient does not have the disease
- Mistake: Test result says “positive” -->> reject H₀, but in reality, H₀ was true

Type II Error Example:
-  A medical test fails to detect a disease in a patient who actually has it
-  H₀: Patient does not have the disease
-  Mistake: Test result says “negative” -->> fail to reject H₀, but in reality, H₀ was false

Key Points:
- Reducing Type I errors often increases Type II errors, and vice versa
- The goal is to balance the risks depending on the situation — for example, in medicine, avoiding Type II errors (missing a disease) is often more important

5.   What is the difference between a Z-test and a T-test? Explain when to use each.
-  Z-test: A statistical test used when the population variance (σ²) is known and the sample size is large (n ≥ 30). It follows the standard normal distribution (Z-distribution)
-  T-test: A statistical test used when the population variance is unknown and the sample size is small (n < 30). It follows the Student’s t-distribution


When to Use Each:
Use Z-test when:
-  Sample size is large (n ≥ 30)
-  Population variance (σ²) is known
-  Data is approximately normally distributed or n is large enough for the Central Limit Theorem to apply

Use T-test when:
-  Sample size is small (n < 30)
-  Population variance (σ²) is unknown
-  Data is approximately normally distributed

| **Aspect**              | **Z-test**                                          | **T-test**                                             |
| ----------------------- | --------------------------------------------------- | ------------------------------------------------------ |
| **Population variance** | Known                                               | Unknown (estimated from sample)                        |
| **Sample size**         | Large (n ≥ 30)                                      | Small (n < 30)                                         |
| **Distribution used**   | Standard normal (Z-distribution)                    | Student’s t-distribution                               |
| **Shape of curve**      | Symmetrical, narrower                               | Symmetrical, slightly wider (heavier tails)            |
| **Precision**           | More precise when σ is known                        | Less precise when σ is unknown, especially for small n |
| **Example use**         | Testing if average height of 500 students is 170 cm | Testing if average height of 15 students is 170 cm     |


Example: 
-  Z-test: A company knows the population standard deviation of battery life is 5 hours. Testing 50 batteries, you check if the mean life is different from 100 hours
-  T-test: A small lab tests 12 new bulbs without knowing the population standard deviation, to see if the average life differs from 800 hours

6.   Write a Python program to generate a binomial distribution with n = 10 and p = 0.5, then plot its histogram.


(Include your Python code and output in the code box below.)


Hint: Generate random number using random function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of random samples

# Generate binomial random numbers
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins = range(n + 2), edgecolor = 'black', align = 'left')
plt.title(f'Binomial Distribution (n = {n}, p = {p})')
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()
```

7.  Implement hypothesis testing using Z-statistics for a sample dataset in Python.


Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


(Include your Python code and output in the code box below.)

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis:
# H0: population mean μ = 50
# H1: population mean μ ≠ 50

# Known population standard deviation (assume for Z-test)
pop_std = 0.5   # assumed known σ

# Parameters
sample_mean = np.mean(sample_data)
n = len(sample_data)
mu_0 = 50  # hypothesized population mean

# Z-statistic formula
z_stat = (sample_mean - mu_0) / (pop_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Output results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision at alpha = 0.05
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: There is significant evidence that μ ≠ 50.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that μ ≠ 50.")
```

8.  Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean.


Plot the data using Matplotlib.


(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mean_true = 50      # true mean of the population
std_true = 5        # true standard deviation
n = 100             # sample size

data = np.random.normal(mean_true, std_true, n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof = 1)  # sample standard deviation
std_error = sample_std / np.sqrt(n)

# Step 3: Calculate 95% confidence interval
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)  # Z-score for 95%
margin_of_error = z_critical * std_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot the data
plt.figure(figsize = (8, 5))
plt.hist(data, bins = 15, edgecolor = 'black', alpha = 0.7)
plt.axvline(sample_mean, color = 'red', linestyle = '--', label = 'Sample Mean')
plt.axvline(ci_lower, color = 'green', linestyle = '--', label = '95% CI Lower')
plt.axvline(ci_upper, color = 'green', linestyle = '--', label = '95% CI Upper')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis = 'y', linestyle = '--', alpha = 0.6)
plt.show()
```

9.   Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Step 1: Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data, ddof = 1)  # sample standard deviation
    
    # Step 2: Calculate Z-scores
    z_scores = (data - mean) / std_dev
    
    # Step 3: Print basic info
    print(f"Mean of original data: {mean:.2f}")
    print(f"Standard deviation of original data: {std_dev:.2f}")
    print(f"First 5 Z-scores: {z_scores[:5]}")
    
    # Step 4: Plot histogram of Z-scores
    plt.figure(figsize = (8, 5))
    plt.hist(z_scores, bins = 15, edgecolor = 'black', alpha = 0.7)
    plt.axvline(0, color = 'red', linestyle = '--', label = 'Mean (0)')
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(axis = 'y', linestyle = '--', alpha = 0.6)
    plt.show()
    
    return z_scores

# Example dataset
np.random.seed(0)
sample_data = np.random.normal(50, 5, 100)  # mean=50, std=5, n=100

# Run the function
z_scores = calculate_and_plot_zscores(sample_data)
```
