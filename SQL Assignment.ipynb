1.  Explain the fundamental differences between DDL, DML, and DQL commands in SQL.


Provide one example for each type of command.

1. DDL – Data Definition Language

Definition:
DDL commands are used to define, create, and modify the structure of database objects such as tables, schemas, and indexes.

Key Points:

Affect the database schema (structure).

Changes are automatically committed (cannot be rolled back).

Used by database administrators or designers.

Common Commands: CREATE, ALTER, DROP, TRUNCATE

2. DML – Data Manipulation Language

Definition: DML commands are used to manipulate or modify the data stored in database tables.

Key Points:

Affect the data inside tables, not the structure.

Changes can be rolled back (transactional).

Used for inserting, updating, or deleting records.

Common Commands: INSERT, UPDATE, DELETE

3. DQL – Data Query Language

Definition: DQL commands are used to query or retrieve data from the database.

Key Points:

Do not modify data; only fetch it.

The main focus is on data retrieval and analysis.

Common Command: SELECT

2.  What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.

Definition: SQL constraints are rules applied to table columns that ensure the accuracy, reliability, and integrity of the data stored in a database.
Purpose: To maintain data integrity by preventing invalid or inconsistent data. To enforce business rules directly within the database structure. To ensure relationships between tables remain consistent.

1. PRIMARY KEY Constraint

Description:  Uniquely identifies each record in a table.
Ensures that the column (or combination of columns) contains unique and non-null values.

2. FOREIGN KEY Constraint

Description: Establishes a relationship between two tables.
Ensures that the value in one table (child) corresponds to a valid value in another table (parent). Maintains referential integrity.

3. CHECK Constraint

Description: Ensures that the values in a column meet a specific logical condition.
Used to restrict the range or format of data values.

3.  Explain the difference between LIMIT and OFFSET clauses in SQL. How would you use them together to retrieve the third page of results, assuming each page has 10 records?

Difference between LIMIT and OFFSET Clauses in SQL

In SQL, the LIMIT and OFFSET clauses are used to control the number of rows returned by a query — especially useful when implementing pagination (displaying results page by page).

1. LIMIT Clause

Definition: The LIMIT clause specifies the maximum number of records to return.

2. OFFSET Clause

Definition:

The OFFSET clause tells the database to skip a specific number of rows before starting to return results.

It is used along with LIMIT to fetch results from a specific position in the dataset.

3. Using LIMIT and OFFSET Together for Pagination

If each page contains 10 records, then:

Page 1: OFFSET 0 → rows 1–10

Page 2: OFFSET 10 → rows 11–20

Page 3: OFFSET 20 → rows 21–30

4.   What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.

1. Definition of Common Table Expression (CTE)

A Common Table Expression (CTE) is a temporary, named result set defined within the execution scope of a single SQL statement.
It is created using the WITH keyword and can be referenced later in the same query — similar to a temporary view.

2. Purpose of a CTE

CTEs are mainly used to:

Simplify complex queries (especially those with nested subqueries).

Improve code readability and maintainability.

Allow recursive queries (useful for hierarchical or tree-structured data).

5.   Describe the concept of SQL Normalization and its primary goals. Briefly explain the first three normal forms (1NF, 2NF, 3NF).


Definition of SQL Normalization (2 marks)

Normalization is the process of organizing data in a database to reduce redundancy (duplicate data) and improve data integrity.
It involves dividing large, complex tables into smaller related tables and establishing relationships between them using foreign keys.

2. Primary Goals of Normalization (2 marks)

Eliminate Data Redundancy – Avoid storing the same data in multiple places.
Ensure Data Consistency – Changes made in one place automatically reflect wherever the data is used.
Simplify Maintenance – Easier to update, insert, or delete data without anomalies.
Improve Query Efficiency – Smaller, well-structured tables lead to faster operations.
